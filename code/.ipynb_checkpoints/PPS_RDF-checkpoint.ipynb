```python
import cmeutils
import matplotlib.pyplot as plt
from cmeutils.structure import gsd_rdf
import hoomd
import numpy as np
```

```python
gsdfile = 'trajectory.gsd'
```

```python
rdf_data, n = gsd_rdf(
    gsdfile,
    'ca',
    's',
    start=-30,
    stop=-1,
    r_max=4,
    r_min=0,
    bins=100,
    exclude_bonded=True,
)
```

```python
rdf_data.bin_centers
```

```python
rdf_data.rdf
```

```python
fig = plt.plot(rdf_data.bin_centers,rdf_data.rdf*n)
plt.ylabel('g(r) (RDF values)')
plt.xlabel('r, in angstroms')
plt.title("RDF of PPS")
plt.show()
```
